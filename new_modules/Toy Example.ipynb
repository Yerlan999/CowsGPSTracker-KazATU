```python
# Function to determine paddock status (With Cattle or Empty)
def paddock_status(load, green_mass, paddock_green_masses):
    if load > 0:
        status_load = 1
    else:
        status_load = 0

    if green_mass <= 0:
        status_resource = -1  # Bad condition
    elif green_mass == max(paddock_green_masses):
        status_resource = 1  # Maximum compared to others
    else:
        status_resource = 0  # Otherwise

    return status_load, status_resource
```

```python
# Iterate through each day
for day in range(len(df)):
    # Extract green masses for each paddock for the current day
    paddock_green_masses = [df[f'Paddock{i}_Green_Mass'].iloc[day] for i in range(1, 8)]
    
    action_needed = False
    # Iterate through each paddock
    for paddock in range(1, 8):
        load_column = f'Paddock{paddock}_Load'
        green_mass_column = f'Paddock{paddock}_Green_Mass'
        status_column = f'Paddock{paddock}_Status'
        
        # Assign status for the paddock
        assess_load, assess_resource = paddock_status(df[load_column].iloc[day], df[green_mass_column].iloc[day], paddock_green_masses)
        if assess_load == 1 and assess_resource == -1: action_needed = True
        df.at[day, status_column] = assess_resource
    
    df.at[day, "Action_Needed"] = int(action_needed)
```

```python
# Function to apply conditions and create 'State' column
def get_state(row):
    if row['Resource'] <= 0 and row['Load'] != 0:
        return 0
    elif row['Resource'] > 0 and row['Load'] != 0:
        return 1
    elif row['Resource'] <= 0 and row['Load'] == 0:
        return 2
    elif row['Resource'] >= 0 and row['Load'] == 0:
        return 3
```

```python
# Sample DataFrame
data = {'Resource': [-1.5, 0.0, 2.0, -3.0, 1.0],
        'Load': [3, 0, 5, 2, 0]}

df = pd.DataFrame(data)
df
```

```python
# Apply the function to create the 'State' column
df['State'] = df.apply(get_state, axis=1)
df
```
