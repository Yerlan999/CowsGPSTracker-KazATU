```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
analysis = pd.read_excel('ground_data2.xlsx', header=None).drop(7,0)
analysis
```

```python
zagons = ["Загон №1", "Загон №2", "Загон №3", "Загон №4", "Загон №5", "Загон №6", "Загон №7"]

strav1 = {"До": analysis.iloc[:,0], 
         "После": analysis.iloc[:,1]}

strav2 = {"До": analysis.iloc[:,3], 
         "После": analysis.iloc[:,4]}

fig, ax = plt.subplots(1,2, figsize=(12,5), sharey=True)

for strav, data in strav1.items():
    p = ax[0].bar(zagons, data, label=strav)
    
    ax[0].bar_label(p)
    
for strav, data in strav2.items():
    p = ax[1].bar(zagons, data, label=strav)
    
    ax[1].bar_label(p)
       
ax[0].set_title('Стравливание №1')
ax[0].legend()
ax[1].set_title('Стравливание №2')
ax[1].legend()
fig.suptitle('Урожайность')
fig.tight_layout()
plt.show()
```
